\begin{array}{l}{{A}^{-1}}=\frac{1}{\left| A \right|}\left( Adj.\,A \right)\end{array}

These are special set of scalars associated with a system of linear equations, most commonly used in matrix equations. The eigenvalue is a scalar that is used to transform the eigenvector. ### **$**2.12. Correlation and covariance matrix

A correlation matrix and a covariance matrix are two key concepts in statistics and data analysis. A correlation matrix is a table that displays the correlation coefficients between different variables. On the other hand, a covariance matrix is a square matrix that provides the covariance between each pair of elements of a given random vector. The variance of each element is represented along the matrix’s major diagonal, and the covariance is represented among the non-diagonal elements. These matrices have wide applications in various fields such as finance, data analysis, and time series analysis.

```r
rown = c("c1","c2")
coln = c("v1","v2","v3","v4")

mat = matrix(c(90,50,46,44,78,65,12,35), byrow = TRUE, nrow = 2,dimnames = list(rown,coln))
print(mat)
print("Correlation matrix of the given marix")
print(cor(mat))
print("Covariance matrix of the given matrix")
print(cov(mat))
```

### **$**2.13. Root of Quadratic equation

```r
# Constructing Quadratic Formula
quadraticformula = function(a,b,c){
  print(paste0("Quadratic equation is: ", a , "x^2 + ", b , "x +", c , " = 0"))
  delta = function(a,b,c){
    return(b^2-4*a*c)
  }
  if(delta(a,b,c) > 0){
    x_1 = (-b+sqrt(delta(a,b,c)))/(2*a)
    x_2 = (-b-sqrt(delta(a,b,c)))/(2*a)
    return(paste0("The two roots are: ", format(round(x_1,2))," and ",format(round(x_2,2))))
  }
  else if(delta(a,b,c) == 0){
    x = -b/(2*a)
  }
  else {"There are no real roots."}
}

x = quadraticformula(1,4,3)
print(x)
```

```r
# Setting seed for reproducibility
set.seed(123)
data = sample(1:20, 1000, replace = TRUE)
```

*  **Simple frequency distribution**

```r
print(summary(data))
```

```r
#Calculating class or group length
print((20-1)/5)
```

```r
breaks_ = seq(min(data), max(data) + 1, by = 5)
grouped_freq_ = cut(data, breaks_, right = FALSE)
grouped_freq_table_ = table(grouped_freq_)
print("Grouped Frequency Distribution:")
print(grouped_freq_table_)
```

```r
# Visualizing grouped frequency distribution for the dataset
barplot(grouped_freq_table_,
 main = "Grouped Frequency Distribution",
 xlab = "Group", ylab = "Frequency",col = c(2:5))
```

```r
# Calculate cumulative frequency for the dataset
cumulative_freq = cumsum(simple_freq_large)
print("Cumulative Frequency Distribution:")
print(cumulative_freq)
```

```r
# Visualizing cumulative frequency distribution
plot(cumulative_freq, type = "o",
main = "Cumulative Frequency Distribution",
xlab = "Values", ylab = "Cumulative Frequency",
col = "blue", lwd = 2)
```

* Hypothesis-

```r
linear_model = lm(y ~ X, data = df)

print(summary(linear_model))
```

Clearly p vlue is very less than 0.05 so H0 is rejected i.e. there is a relationship between X and y; where our model explain data by 60%

```r
# Model prediction
prediction = predict(linear_model)
rss = resid = (y - prediction)

df1 = data.frame(as.matrix(df),prediction = as.vector(prediction),rss = as.vector(rss))
print(head(df1))
```

**Assessing the accuracy of the model**

```r
rse = round(sqrt(sum((df1$rss)^2)/(length(df1$rss)-2)),3)
F_Statistic = round(summary(linear_model)$fstatistic[1],3)
R_2 = round(summary(linear_model)$r.squared,3)

accuracy = data.frame(RSE = rse, F_Statistic = F_Statistic, R_2_in_percentage = R_2*100)
print(accuracy)
```

```r
# Checking the importance of predictors
print(anova(lr))
```

All p-values are less than 0.05 so we accept our H1.

Clearly all features are important since all p-value is less than 0.05.

Clearly there is no multicolinearity.

          H0: rho = 0
          H1: rho != 0

* p-value greater than 0.05, so do not reject H0 that this data set not a time series data
* So we go for cross sectional reg

```r
auto_corr = 1 - 0.5 * dwt(lr)$dw
print(auto_corr)
```

### **$**6.1. For Univariate Population Hypothesis Testing

**Our Hypothesis for test of mean**

          H0: mean = 50
          H1: mean > 50
              or
          H1: mean != 50

* **Performing t test one sided**


Clearly H0 is rejected on the basis of p-value (p=value is less than 0.05) and we accept alternative.

```r
one_t = t.test(mu = 50,
              df_asia$lifeExp,
              conf.level = 0.95,
              alternative = "greater")
print(one_t)
```

* **Confidance interval at 95%**

```r
print(one_t$conf.int[c(1,2)])
```

* **Two sided t-test**

p-value is less than 0.05 so we reject our H0 and accept alternative.

```r
two_t = t.test(mu = 50, df_asia$lifeExp, conf.level = 0.99)
print(two_t)
```

* **Confidance interval at 99%**

```r
print(two_t$conf.int[c(1,2)])
```

**Test for variance**

      H0: sigma = 11
      H1: sigma != 11

```r
sigma = 11
# Calculate the chi-square statistic
chi_square_stat = sum((df_asia$lifeExp - mean(df_asia$lifeExp))^2) / sigma^2

# Degrees of freedom
n = length(df_asia$lifeExp) - 1

# Calculate p-value
p_value = 1 - pchisq(chi_square_stat, n)

# Print the results
cat("Chi-Square Statistic:", chi_square_stat, "\n")
cat("Degrees of Freedom:", n, "\n")
cat("p-value:", p_value, "\n")
```

p-value is less than 0.05 so H0 rejected our alternative accepted.
